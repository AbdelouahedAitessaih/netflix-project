```python
import pandas as pd

df = pd.read_csv("D:/EMG/2eme annee/S2/Programmation python/Projets/netflix-project/data/raw/netflix_shows.csv")

df.head()
```

```python
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["genres"] = df["genres"].fillna("Unknown")
df["description"] = df["description"].fillna("Unknown")
df["budget"] = df["budget"].fillna("Unknown")
df["revenue"] = df["revenue"].fillna("Unknown")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
```

```python
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
```

```python
df["genres"] = df["genres"].str.split(", ")
```

```python
import ast

df["genres"] = df["genres"].apply(ast.literal_eval)
```

```python
type(df["genres"].iloc[0])
```

```python
df["type"] = df["type"].str.lower()
```

```python
df["budget"] = pd.to_numeric(
    df["budget"], errors="coerce"
)

df["revenue"] = pd.to_numeric(
    df["revenue"], errors="coerce"
)
```

```python
df = df.drop_duplicates()
```
