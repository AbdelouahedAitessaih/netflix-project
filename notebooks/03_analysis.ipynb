# Analyse pour le dashboard Tableau

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "D:/EMG/2eme annee/S2/Programmation python/Projets/netflix-project"

df = pd.read_csv(f"{path}/data/processed/netflix_clean.csv")
genres = pd.read_csv(f"{path}/dashboard/tableau_export/netflix_genres.csv")
pays = pd.read_csv(f"{path}/dashboard/tableau_export/netflix_pays.csv")
```

## Genres populaires

```python
top_genres = genres["genres"].value_counts().head(10)
top_genres
```

```python
top_genres.plot(kind="barh")
plt.title("Genres les plus frequents")
plt.show()
```

## Evolution des sorties

```python
par_annee = df["year_added"].value_counts().sort_index()
par_annee
```

```python
par_annee.plot(kind="line", marker="o")
plt.title("Nombre de titres ajoutes par annee")
plt.show()
```

## Analyse regionale

```python
top_pays = pays["country"].value_counts().head(10)
top_pays
```

```python
top_pays.plot(kind="bar")
plt.title("Top 10 pays")
plt.xticks(rotation=45)
plt.show()
```
