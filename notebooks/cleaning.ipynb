# Nettoyage des donnees

```python
import pandas as pd

path = "D:/EMG/2eme annee/S2/Programmation python/Projets/netflix-project"

df = pd.read_csv(f"{path}/data/raw/netflix_shows.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["genres"] = df["genres"].fillna("Unknown")
df["description"] = df["description"].fillna("Unknown")
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
```

```python
df["type"] = df["type"].str.lower()
df = df.drop_duplicates()
df = df[df["title"] != "Unknown"]
df = df[df["genres"] != "Unknown"]
```

```python
df = df.drop(columns=["budget", "revenue"], errors="ignore")
df.info()
```

```python
genres = df["genres"].str.split(", ").explode()
genres.value_counts().head(10)
```

```python
df.to_csv(f"{path}/data/processed/netflix_clean.csv", index=False)
df.to_csv(f"{path}/dashboard/tableau_export/netflix_clean.csv", index=False)
```

```python
# export pour tableau - genres
genres_df = df[["show_id", "title", "type", "genres", "year_added", "rating"]].copy()
genres_df["genres"] = genres_df["genres"].str.split(", ")
genres_df = genres_df.explode("genres")
genres_df.to_csv(f"{path}/dashboard/tableau_export/netflix_genres.csv", index=False)
```

```python
# export pour tableau - pays
pays_df = df[["show_id", "title", "type", "country", "year_added"]].copy()
pays_df["country"] = pays_df["country"].str.split(", ")
pays_df = pays_df.explode("country")
pays_df.to_csv(f"{path}/dashboard/tableau_export/netflix_pays.csv", index=False)
```

```python
# evolution des sorties par annee et type (pour Tableau)
sorties = df.groupby(["year_added", "type"]).size().reset_index(name="count")
sorties.to_csv(f"{path}/dashboard/tableau_export/sorties_annuelles.csv", index=False)
sorties.tail(10)
```
