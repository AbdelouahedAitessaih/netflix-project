```python
import pandas as pd

df = pd.read_csv("D:/EMG/2eme annee/S2/Programmation python/Projets/netflix-project/data/processed/netflix_clean.csv")
df.head()
```

```python
df["type"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["type"].value_counts().plot(kind="bar")
plt.title("Répartition Movies vs TV Shows")
plt.show()
```

```python
df["year_added"].value_counts().sort_index()
```

```python
df["year_added"].value_counts().sort_index().plot(kind="line")
plt.title("Évolution du contenu Netflix par année")
plt.show()
```

```python
df["country"].value_counts().head(10)
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 pays producteurs Netflix")
plt.show()
```
