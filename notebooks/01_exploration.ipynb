# Exploration Netflix

```python
import pandas as pd
import matplotlib.pyplot as plt

path = "D:/EMG/2eme annee/S2/Programmation python/Projets/netflix-project"

df = pd.read_csv(f"{path}/data/processed/netflix_clean.csv")
df.head()
```

```python
df.info()
```

```python
df["type"].value_counts()
```

```python
df["type"].value_counts().plot(kind="bar")
plt.title("Films vs series")
plt.show()
```

```python
df["year_added"].value_counts().sort_index()
```

```python
df["year_added"].value_counts().sort_index().plot(kind="line")
plt.title("Evolution des ajouts par annee")
plt.show()
```

```python
df["country"].value_counts().head(10)
```

```python
df["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 pays")
plt.xticks(rotation=45)
plt.show()
```

```python
genres = df["genres"].str.split(", ").explode()
genres.value_counts().head(10)
```

```python
genres.value_counts().head(10).plot(kind="bar")
plt.title("Top genres")
plt.xticks(rotation=45)
plt.show()
```
